# EM-алгоритм

В этом примере мы рассмотрим, что из себя представляет EM-алгоритм (Expectation–maximization algorithm). Затем продемонстрируем его в действии на примере задачи кластеризации точек на плоскости.

## Краткое описание

В задаче восстановления плотности распределения требуется апроксимировать плотность распределения входных данных. Иногда искомое распределение плохо описывается стандартными распределениями, либо при описании одним параметрическим семейством распределений возникает слишком много параметров, которые очень легко переобучить. 

В таком случае можно попробовать представить распределение как смесь распределений из каких-то известных семейств:
$$p(x) = \sum_{j=1}^k w_j p_j(x),\space\space\space\space \sum_{j=1}^k w_j = 1,\space\space\space\space w_j > 0,\space\space\space\space x \in X^m,$$
где $w_j$ - априорная вероятность $j$-й компоненты смеси, $p_j(x)$ - её плотность распределения, $X^m$ - обучающая выборка.

Как правило полагают, что распределения всех компонент принадлежат одному семейству, т.е. отличаются лишь параметрами: $p_j(x) = \varphi(x, \theta_j)$

Задача разделения смеси распределений заключается в поиске априорных вероятностей и вектора параметров $\Theta = \{w_1, ..., w_k, \theta_1, ..., \theta_k\}$. Рассмотрим EM-алгоритм, решающий эту задачу.

Вводим вспомогательный вектор "скрытых" переменных $G$ такой, что:
- он легко считается из $\Theta$
- из него легко апроксисмируется $\Theta$ (методом максимального правдоподобия)

Это позволит улучшать оценку $\Theta$ методом "перегона" в $G$ и обратно. Тогда EM-алгоритм будет выглядеть так:
- $\Theta = \Theta_0 $ // начальное приближение
- пока $\Theta$ и $G$ не стабилизируются:
    - $G$ = ExpectationStep($\Theta$)
    - $\Theta$ = MaximizationStep($G$)

<b>ExpectationStep:</b>

Положим:
$$g_{ij} = \frac{w_i p_j(x_i)}{\sum_{s=1}^k w_s p_s(x_i)} \space\space\space\space \forall i,j,\space\space\space\space  G = (g_{ij})_{m \times k} $$

<b>MaximizationStep:</b>

Методом максимального правдоподобия (отыскивая нули производной от логарифма функции правдоподобия) для каждой компоненты смеси вычислим:

$$w_j = \frac{1}{m} \sum_{i=1}^m g_{ij}, \space\space\space\space \theta_j = arg \max_{\theta} \sum_{i=1}^m g_{ij} \ln \varphi(x_i, \theta), \space\space\space\space \forall j \in [1,k]$$

<b>Когда заканчивать:</b>

Имеет смысл остановиться, когда компоненты $\Theta$ перестают изменяться сильно. Дальнейшие изменения не окажут особого влияния на точность.

<b>Как выбрать начальные приближения:</b>

EM-алгоритм сходится к локальному экстремуму. От выбора начальных приближений сильно зависит качество такого решения, поэтому здесь используют эвристический подход.
Стандартная (но далеко не самая лучшая) эвристика - выбрать компоненты случайно. Другая эвристика - выбрать центры компонент максимально удаленными друг от друга. Лучше всего запускать алгоритм несколько раз с разными начальными данными и выбирать лучший. Также часто сначала пробуют кластеризовать точки каким-нибудь простым алгоритмом кластеризации (например, kMeans) и затем использовать параметры, вычисленные в каждом полученном кластере (параметрическим восстановлением плотности распределения) как начальные приближения для EM-алгоритма.

<b>Где используется EM-алгоритм:</b>
    
Мы можем оценивать плотность распределения как смесь распределений, но это можно делать с разными целями. Вот три самые частые:
- оценить как смесь распределений плотность $p(x|y)$ класса $y$ для каждого класса в задаче классификации;
- оценить как смесь распределений плотность $p(x)$ всех точек из разных классов в задаче классификации, а параметры смеси интерпретировать как априорные вероятности классов $P(y) = w_y$ и плотности распределений классов $p(x|y) = p_y(x)$, жестко задав в EM-алгоритме для точек обучающей выборки $p(y|x_i) = g_{iy} = [y = y_i]$ (единица для правильного класса, ноль для остальных)
- оценить как смесь распределений плотность $p(x)$ всех точек из разных кластеров в задаче кластеризации, а параметры смеси интерпретировать как априорные вероятности кластеров $P(y) = w_y$ и плотности распределений кластеров $p(x|y) = p_y(x)$ (в этом случае ни для каких точек нет информации о правильном классе, поэтому жестко задавать значения скрытых переменных не нужно)

<b>Где лучше изучить матчасть:</b>

Воронцова <a href="http://www.ccas.ru/voron">"Математические методы обучения по прецедентам".</a> Стоит заметить, что вывод EM-алгоритма в этих лекциях намного проще для понимания, чем классический, но в нем теряется смысл названия E-шага: непонятно, причем здесь "ожидание". Желающие разобраться в этом могут обратиться, хотя бы, к Википедии.


## Кластеризация EM-алгоритмом

Для простоты визуализации рассмотрим случай двух признаков, чтобы объекты были точками на плоскости.
Сгенерируем скопления точек вокруг (4, 0) и (8, 0), которые формируют 2 явно выделенных кластера.

Для этого сконструируем матрицу с координатами точек в строках. Возьмем вектор-столбец высоты 100 из единиц, и умножим на него массив (4, 0):

$$ \begin{pmatrix} 4 & 0 \end{pmatrix} \begin{pmatrix} 1 \\ 1 \\ \cdots \\ 1 \end{pmatrix} = \begin{pmatrix} 4 & 0 \\ 4 & 0 \\ \cdots & \cdots \\ 4 & 0 \end{pmatrix}$$

Остается сложить с матрицей 100х2 из распределенных стандартно нормально (т.е. нормально с матожиданием 0 и дисперисей 1) случайных величин:


```python
import numpy as np
from scipy import *

np.random.seed(1)
points_count = 500

# к матрице из случайно сгенерированных чисел размера 100х2 прибавляем
# (4, 0) * (1, 1, ..., 1) = матрица 100х2, где первая строка состоит из 4 или 8, вторая из нулей
points4 = np.random.randn(points_count, 2) + (array([4, 0]) * ones((points_count, 1)))
points8 = np.random.randn(points_count, 2) + (array([8, 0]) * ones((points_count, 1)))

points = np.concatenate((points4, points8))
```

Покажем обучающую выборку на графике при помощи библиотеки <a href="http://matplotlib.org">matplotlib</a>.

```python
%matplotlib inline
# чтобы изображения выводились прямо в ipython notebook, а не в отдельном окне

import matplotlib.pyplot as plt

# задаем координаты, цвет (b == blue) и формат вывода (o == круглая точка)
plt.plot(points[:,0], points[:,1], 'bo')

# задаем границы видимой части координатных осей
plt.axis([0, 12, -6, 6])

plt.show()
```

Теперь воспользуемся библиотекой <a href="http://scikit-learn.org">scikit-learn</a>. Как раз для случаев, когда мы имеем дело со смесью распределений, в scikit-learn имеется пакет mixture. В классе GMM (Gaussian Mixture Models) этого пакета реализован EM-алгоритм для смеси семейства нормальных распределений.

```python
from sklearn import mixture

# создаем модель с 2мя компонентами
g = mixture.GMM(n_components=2)

g.fit(points)
```

Распределим точки по кластерам (спрогнозируем кластер для каждой):

```python
predictions = g.predict(points)
```

Разделим точки по двум массивам в зависимости от кластера:

```python
cluster0 = []
cluster1 = []
for i in range(2 * points_count):
    if predictions[i] == 0:
        cluster0.append(points[i])
    else:
        cluster1.append(points[i])
```

То же самое, но более "Python-style":

```python
get_cluster = lambda cluster_id: [point for point, cluster in zip(points, predictions) if cluster == cluster_id]

cluster0, cluster1 = get_cluster(0), get_cluster(1)
```

Покажем кластеры на графике, покрасив первый кластер в синий цвет, второй - в красный

```python
cluster0 = array(cluster0)
cluster1 = array(cluster1)

plt.plot(cluster0[:,0], cluster0[:,1], 'bo')
plt.plot(cluster1[:,0], cluster1[:,1], 'ro')
plt.axis([0, 12, -6, 6])
plt.show()
```

Конечно нам теперь хочется посмотреть на параметры восстановленной смеси. Точек, конечно, мало, но в идеальном случае мы должны увидеть веса $w_1 = w_2 = \frac{1}{2}$ и единичные дисперсии по обеим координатам у обоих кластеров. 

```python
print "Веса компонент:", g.weights_
print "Матрицы ковариаций:", g.covars_
```

Для 100 точек в кластере результат мало похож на ожидаемый, но уже для 500 точек получаем практически то, что надо.

## Восстановление плотности

Хорошо, а если нам нужно применить EM-алгоритм чтобы просто оценить плотность, а не для кластеризации? Для этого у класса GMM есть метод score:

```python
print g.score([[4, 0], [8, 0]])
```

Полученные числа - это логарифмы значений плотности восстановленной смеси распределений в заданных точках. Попробуем построить heat map для восстановленной смеси - раскрасить точки плоскости в зависимости от значения плотности (чем больше значение, тем "горячее" точка).

```python
from numpy import exp,arange
from pylab import meshgrid,cm,imshow,contour,clabel,colorbar,axis,title,show

# Определим функцию, вычисляющую плотность в каждой точке по массиву координат x и массиву координат y точек
def z_func(x,y):
    t = g.score(zip(x, y))
    return exp(t)
 
x = arange(0.0, 12.0, 0.5)
y = arange(-6.0, 6.0, 0.5)
X,Y = meshgrid(x, y) # сетка точек
```

Посмотрим, как выглядит сетка:

```python
print X[:5]
```

```python
print Y[:5]
```

```python
Z = [z_func(x, y) for x, y in zip(X, Y)] # вычисляем значения плотности на сетке

im = imshow(Z,cmap=cm.hot) # показываем изображение (cmap используем, чтобы задать цветовую схему)

colorbar(im) # добавляем colorbar

# и заголовок
title('Gaussian Mixture $p(x) = w_1 p_1(x) + w_2 p_2(x)$ ')
show()
```

Сравним полученное изображение с построенной по тем же точкам двумерной гистограммой (простейший случай непараметрического оценивания плотности)

```python
import numpy as np
import numpy.random
import matplotlib.pyplot as plt

# Выбираем только координаты x и y из массива с точками
x = [point[1] for point in points]
y = [point[0] for point in points]

# Строим двумерную гистограмму по точкам 
# bins - количество клеток гистограммы по каждой оси
# Итоговые цвета сглаживаются
heatmap, xedges, yedges = np.histogram2d(x, y, bins=20)
extent = [yedges[0], yedges[-1], xedges[0], xedges[-1]]

plt.clf()
plt.imshow(heatmap, extent=extent, cmap=plt.get_cmap('hot'))
plt.show()
```

Как и следовало ожидать, видим две крайности: у смеси получилась очень гладкая и идеализированная картинка, а гистограмма дает очень "скачущую" оценку плотности, не слишком пригодную на практике.

## Классификация

Рассмотрим теперь следующую ситуацию: мы знаем, какие у нас должны быть классы. Но классы неоднородные. Например, мы классифицируем изображения символов и один и тот же символ (например, цифра "4") может выглядеть совсем по-разному:

<img src="files/graphemes.jpg" />

В этой ситуации особенно уместно представить класс как смесь нескольких компонент и оценивать плотность каждого класса отдельно для него запуская EM-алгоритм. Фактически мы можем кластеризовать каждый класс и сравнивать новые объекты уже не с классом в целом, а с каждым кластером - с каким-то характерным начертанием символа. Так возникает кластеризация в задаче выделения графем.

### Задача определения графем

Графема - графическая единица письма, начертание буквы (символа) или лигатуры. Один и тот же символ может изображаться по-разному: ему соответствуют несколько разных графем. Пример: a и a.
База эталонов обычно содержит по несколько эталонов для каждой графемы, каждый из которых называется кластером.
Кластеризация позволяет избежать «размывания» эталона и сделать его более избирательным.
Классификатор в задаче фиксированный. Для каждого изображения считаются расстояния до всех кластеров по формуле:
$$w_0 = \sum_{i=1}^{N} w_i (x_i - m_i)^2$$
Здесь $x_i$ - значение i-того признака, $m_i$ - среднее значение $i$-того признака в кластере, $w_i = \frac{1}{D_i}$, где $D_i$ - дисперсия i-того признака.

### Оценка качества

Описываемый классификатор применяется для быстрого прогноза с некоторым приемлемым качеством, поэтому ему нужно находить несколько наиболее правдоподобных классов, которые затем будут отправлены в более точный классификатор. В данной задаче правильным распознаванием у нас считается попадание правильного варианта в топ-4.
