```python
import numpy as np
from matplotlib import pylab as plt
import scipy.optimize as opt
%matplotlib inline
```

```python
def poly(x, a):
    return [a[0] * i**3 + a[1] * i**2 + a[2] * i + a[3] for i in x]

def poly_coef(x, a, b, c, d):
    return [a * i**3 + b * i**2 + c * i + d for i in x]
```

```python
np.random.seed(123)
x = np.linspace(-2, 2, 100)
y = poly(x, [0.5, 1, -1, 3.5]) + 0.1*np.random.randn(100)
print y
```

```python
plt.plot(x,y)
plt.show()
```

```python
y_pos = np.random.rand(10)*100
for i in y_pos:
    y[round(i)] = y[i] + np.random.randn(1)*10
print y
```

```python
model = opt.curve_fit(poly_coef, x, y)
coef = model[0]
```

```python
y_new  = poly(x, coef)
plt.plot(x, y, label = r'y generated')
plt.plot(x, y_new, label = r'y predicted')
plt.legend()
plt.show()
```

```python
def quality_func(coef):
    res = 0
    y_my = poly(x, coef)
    for i in range(len(y)):
        res += abs(y_my[i] - y[i]) ** 2
    return res

print quality_func([1.0, 1.0, 1.0, 1.0])
```

```python
quality = opt.minimize(quality_func, [1.0, 1.0, 1.0, 1.0])
coef_opt = quality.x
print coef_opt
```

```python
y_new_opt = poly(x, coef_opt)
plt.plot(x, y, label = r'y generated')
plt.plot(x, y_new, label = r'y predicted')
plt.plot(x, y_new_opt, label = r'y predicted 2')
plt.legend()
plt.show()
```
