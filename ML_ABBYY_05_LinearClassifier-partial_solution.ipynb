### 2. Реализация решающей функции

Решающая функция в случае линейного классификатора устроена очень просто: $f(x) = w^T x = \sum\limits_{k = 1}^{n} w_k x_k$, если $f(x) > w_0$ возвращаем класс 1, иначе - 0. Вектор весов признаков $w$ и порог $w_0$ $-$ параметры алгоритма, которые настраиваются при обучении. С обычными типами данных реализовать совсем просто, но у вас $x$ $-$ строка в разреженной(!) матрице признаков. Поэтому несколько комментариев к работе с разреженными матрицами.

```python
print X
```

Как видите, библиотека scipy, в которой реализованы разреженные матрицы, заботливо уберегла нас от того, чтобы видеть всю матрицу. В то же время понятно, как хранятся значения из нее: для каждой пары (строка, столбец), где значение ненулевое, хранится само значение и номера строки и столбца.

Обратите внимание, что сначала приведены ненулевые элементы нулевой строки, потом первой и так далее. Это особенность формата CSR. Есть также формат CSC - там значения группируются по номерам столбцов. От выбора формата матрицы зависит, какие операции с ней можно производить быстро. Если в процессе выполнения задания вы будете получать warning с предложением изменить формат матрицы, значит что-то вы делаете не так: формат CSR замечательно подходит для линейных классификаторов.

Теперь, чтобы реализовать разреженную функцию, остается только придумать, как умножать неразреженный вектор весов признаков на строку из разреженной матрицы признаков, т.е. заменить "w * x" в коде ниже на то, что будет учитывать разреженность x.

```python
def decision_function(x, w, w0):
    return (w * x.transpose())[0, 0] - w0 # нужно заменить w * x

print decision_function(X[0], X[0], 1.0)
```

### 3. Реализация функционала риска

Параметры классификатора подбираются так, чтобы минимизировать функционал риска. Значит, нам остается написать функционал и вызвать для него какой-нибудь метод из scipy.optimize

Для начала определим функцию потерь (функционал риска - просто сумма значений функции потерь на каждом объекте обучающей выборки)

```python
from math import exp, log

def logistic_loss(margin):
    return log(1.0 + exp(-margin))
```

Проверим, что как-то работает:

```python
print logistic_loss(1)
```

Теперь остается вспомнить, что такое отступ (margin). Т.к. в данном случае классы 0 и 1, а мы давали определение для -1 и +1, отступ на объекте $x$ при правильном ответе $y \in \lbrace 0; 1 \rbrace$ вычисляется как $(2y - 1) \cdot$ decision_function($x$)

```python

def risk(w, w0, X_train, y_train):
    '''
    Код, пробегающийся по объектам обучающей выборки, вычисляющий на каждом из них отступ и значение функции потерь,
    и возвращающий сумму значений функции потерь на всех отступах.
    
    '''
    q = 0
    for x, ans in zip(X_train, y_train):
        print "\n\n\n"+str(type(ans))+"\n\n\n"
        margin = (2.0 * ans - 1) * decision_function(x, w, w0)
        loss = logistic_loss(margin)
        q += loss
        
    return q
```
